DISPLOT: distributions(histplot, kdeplot, ecdfplot, rygplot)

```python
import seaborn as sns
```

```python
car_crashes = sns.load_dataset("car_crashes")
car_crashes.head()
```

```python
sns.relplot(x="total", y="speeding", hue="alcohol", data=car_crashes)
```

```python
sns.relplot(x="total", y="speeding", hue="alcohol",data=car_crashes)
#sns.catplot(x="total", y="speeding", kind="bar")
```

```python
sns.catplot(x="total", kind="count", hue="speeding", data=car_crashes)
```
